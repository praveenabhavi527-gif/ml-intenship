```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
stu=pd.read_csv("/content/drive/MyDrive/student_exam_data.csv")
stu.head()
```

```python
stu.isnull().sum()
```

```python
ind=stu[['Study Hours','Previous Exam Score']]
dep=stu['Pass/Fail']
```

```python
Logr=LogisticRegression()
```

```python
Logr.fit(ind,dep)
```

```python
Study_Hours=int(input("Enter the Study hours"))
Previous_Exam_Score=int(input("Enter the Previous Exam Score"))
pred=Logr.predict([[Study_Hours,Previous_Exam_Score]])
print(pred)
```

```python
from sklearn.metrics import accuracy_score
pval=Logr.predict(ind)
accuracy_score(dep,pval)
```
