```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pickle # Import the pickle library

# 1. Load your data
# Note: Ensure your CSV file path is correct
insdf = pd.read_csv('/content/drive/MyDrive/simplelinearregression.csv')

# 2. Prepare the data
inp = insdf[['Age']]
out = insdf['Premium']

# 3. Train the model
LR = LinearRegression()
LR.fit(inp, out)

# 4. Make the prediction (as shown in your screenshot)
prediction = LR.predict([[30]])
print(f"Prediction for Age 30: {prediction}")

# ==================================================
# CODE TO CONVERT/SAVE AS PICKLE FILE
# ==================================================

# Define the filename
pickle_filename = 'linear_regression_model.pkl'

# Open the file in 'write binary' (wb) mode and dump the model
with open(pickle_filename, 'wb') as file:
    pickle.dump(LR, file)

print(f"Success! Model saved as {pickle_filename}")
```
