```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Market_Basket_Optimisation.csv', header=None)
```

```python
transactions = []
for i in range(len(dataset)):
    transactions.append([str(dataset.values[i, j])
                         for j in range(dataset.shape[1])
                         if str(dataset.values[i, j]) != 'nan'])
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df,
                            min_support=0.003,
                            use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets,
                           metric="lift",
                           min_threshold=1)
```

```python
rules = rules.sort_values(by='lift', ascending=False)
rules.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(rules['support'], rules['lift'],
            c=rules['confidence'], cmap='viridis', s=50)

plt.colorbar(label='Confidence')
plt.xlabel('Support')
plt.ylabel('Lift')
plt.title('Apriori Rules: Support vs Lift')
plt.show()
```

```python
top_rules = rules.head(10)

plt.figure(figsize=(10,5))
plt.bar(range(len(top_rules)), top_rules['lift'])
plt.xticks(range(len(top_rules)),
           top_rules['antecedents'].astype(str),
           rotation=90)
plt.ylabel('Lift')
plt.title('Top 10 Association Rules by Lift')
plt.show()
```
