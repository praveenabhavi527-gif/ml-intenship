```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dia=pd.read_csv('/content/drive/MyDrive/diabetcsv.csv')
```

```python
dia.isnull().sum()
```

```python
ind=dia[['age','mass','insu','plas']]
dep=dia['class']
```

```python
Logr=LogisticRegression()
Logr.fit(ind,dep)
```

```python
age=int(input("Enter the age:"))
mass=int(input("Enter the mass:"))
insulin=int(input("Enter insulin level:"))
plasma=int(input("Enter plasma glucose:"))
pred=Logr.predict([[age,mass,insulin,plasma]])
print(pred)
```

```python
Logr.score(ind,dep)
```

```python
from sklearn.metrics import accuracy_score
pval=Logr.predict(ind)
accuracy_score(dep,pval)
```

```python
import pickle

# This name can be whatever you want (e.g., 'saved_model.pkl')
filename = 'diabetes_model.pkl'

# 'wb' opens the file in "write binary" mode, which is required for pickling
with open(filename, 'wb') as file:
    pickle.dump(Logr, file)

print(f"Model saved successfully as {filename}")
```
